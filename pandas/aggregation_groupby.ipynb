# Aggegation y groupby

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
pd.options.display.float_format = '{:,.3f}'.format
```

```python
df = sns.load_dataset('tips')
df
```

```python
df.describe(include="all")
```

```python
df['day'].value_counts() 
```

```python
df['day'].value_counts() / df['day'].value_counts().sum() * 100
```

```python
df.groupby('sex').mean()
```

```python
df['prct_tip'] = df['tip'] / df['total_bill']
df
```

```python
df.groupby('sex').mean()
```

```python
df.groupby('sex').median()
```

```python
df.groupby('sex')[['total_bill','prct_tip']].describe().T
```

```python
def mean_eur2usd(x):
    return np.mean(x)*1.12
```

```python
mean_eur2usd(100)
```

```python
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(mean_eur2usd)
```

```python
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(np.std)
```

```python
df.groupby(['sex','time'])[['total_bill','prct_tip']].apply(lambda x: np.mean(x)*1.12)
```

```python
df.groupby(["sex","time"])[["total_bill","prct_tip"]].aggregate([np.mean, np.max, np.std])
```

```python
df.groupby(["sex","time"])[["total_bill","prct_tip"]].agg([np.mean, np.max, np.std])
```

```python
dict_agg = {
    'tip': [min, max],
    'total_bill': [np.mean, mean_eur2usd]}

dict_agg
```

```python
df.groupby(["sex","time"])[["total_bill","tip"]].agg(dict_agg)
```

```python
def f_filter(x):
    return mean_eur2usd(x["total_bill"].mean()) > 20
```

```python
df_filtered = df.groupby(["sex","time"]).filter(f_filter)
df_filtered
```

```python
df_filtered.groupby(["sex","time"]).count()
```

## Pivot Table

```python
df
```

```python
df.groupby(["sex","time"])[["total_bill"]].mean()
```

```python
df_gp = df.groupby(["sex","time"])[["total_bill"]].mean().reset_index()
```

```python
df_gp
```

```python
df_gp.pivot_table(values="total_bill", index="sex",columns="time")
```

```python
df_gp.pivot_table(values="total_bill", index="sex",columns="time", aggfunc=np.median)
```

```python
df_gp.pivot_table(values="total_bill", index="sex",columns="time", aggfunc=[np.median, np.std])
```
