```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use("classic")
image = Image.open("./imagenes/zebra.jpg")
plt.imshow(image)
```

```python
image_gr = image.convert("LA")
image_mat = np.array(list(image_gr.getdata(band=0)), float)
image_mat.shape = (image_gr.size[1], image_gr.size[0])
```

```python
U, D, V = np.linalg.svd(image_mat)
```

```python
print(image_mat.shape)
```

```python
print(U.shape)
print(D.shape)
print(V.shape)
```

```python
img_rec = np.matrix(U[:,:1]) * np.diag(D[:1]) * np.matrix(V[:1,:])
plt.imshow(img_rec, cmap="gray")
```

```python
i = 5
img_rec = np.matrix(U[:,:i]) * np.diag(D[:i]) * np.matrix(V[:i:])
plt.imshow(img_rec, cmap="gray")
title = "Valores singulares = %s" % i
plt.title(title)
plt.show()
```
