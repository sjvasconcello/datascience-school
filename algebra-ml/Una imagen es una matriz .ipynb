```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use("classic")

img = Image.open("./imagenes/zebra.jpg")

plt.imshow(img)
```

```python
img_gr = img.convert("LA")
print(img_gr)
```

```python
img_mat = np.array(list(img_gr.getdata(band=0)), float)
print(img_mat)
```

```python
img_mat.shape = (img_gr.size[1], img_gr.size[0])
print(img_mat)
```

```python
img_mat.shape
```

```python
plt.imshow(img_mat, cmap = "gray")
```

```python
imagen_mat_2 = img_mat / 10
plt.imshow(imagen_mat_2)
```

```python
plt.imshow(imagen_mat_2, cmap="gray")
```
