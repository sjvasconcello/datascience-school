```python
#Llamamos el set de datos público 
zoo_data = 'https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.data'

zoo_name= ['animal name','hair','feathers','eggs',
           'milk','airborne','aquatic','predator',
           'toothed','backbone','breathes','venomous',
           'fins','legs','tail','domestic',' cat-size','type']
```

```python
import pandas as pd
```

```python
df_zoo = pd.read_csv(zoo_data)
df_zoo.head(10)
```

```python
df_zoo.columns= zoo_name
df_zoo.head(10)
```

```python
import altair as alt
```

```python
plot_01 = alt.Chart(df_zoo).mark_line().encode(
x="animal name", y="legs")

plot_01
```

```python
plot_02 = alt.Chart(df_zoo).mark_bar().encode(
x="animal name", y="legs")
plot_02
```
