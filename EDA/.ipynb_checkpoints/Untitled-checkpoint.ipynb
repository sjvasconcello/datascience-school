```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
b_cancer = load_breast_cancer()
```

```python
cancer_data = b_cancer.data
cancer_name = b_cancer.feature_names
```

```python
df_cancer = pd.DataFrame(cancer_data, columns=cancer_name)
df_cancer
```

```python
df_cancer.info()
```

```python
df_cancer_test = df_cancer.iloc[:,:6]
df_cancer_test
```

```python
df_cancer_test.corr()
```

```python
corr_matrix = df_cancer_test.corr()
mask = np.zeros_like(corr_matrix)
mask[np.triu_indices_from(mask)] = True

with sns.axes_style("husl"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr_matrix, mask=mask, vmax=.3, square=True, annot=True)

```
