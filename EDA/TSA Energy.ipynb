```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/df_power.csv"
```

```python
data = pd.read_csv(url)
```

```python
data.head()
```

```python
data.columns
```

```python
data.info()
```

```python
data["Date"] = pd.to_datetime(data["Date"])
data.info()
```

```python
data = data.set_index("Date")
```

```python
data["year"] = data.index.year
```

```python
data.head()
```

```python
data["month"] = data.index.month
data.head()
```

```python
data["day_name"] = data.index.day_name()
data.head()
```

```python
data["Consumption"].plot(linewidth=0.5)
```

```python
data.info()
```

```python
name = ["Consumption","Wind","Solar"]
data[name].plot(marker=".",
                alpha=0.5,
                linestyle="None",
                figsize=(14,6), 
                subplots=True)
```

```python
data.loc["2016","Consumption"].plot()
```

```python
data.loc["2016-12","Consumption"].plot(marker="o")
```

```python
fig, axess=plt.subplots(3,1, figsize=(8,7))
for name, ax in zip(["Consumption","Wind","Solar"],axess):
    sns.boxplot(data=data,x="month",y=name,ax=ax)
```
