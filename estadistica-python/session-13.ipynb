```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/iris-data.csv", index_col=0)
```

```python
df.columns
```

```python
df.tipo_flor.value_counts()
```

```python
y = df["lar.petalo"]
fig, ax = plt.subplots()
ax.set_title("Variable Original")
ax.hist(y, bins=40)
ax.axvline(x=np.mean(y), c="k", label="media", linestyle="--")
ax.axvline(x=np.mean(y) + np.std(y), c="r", label="+1 desviacion estandar", linestyle="--")
ax.axvline(x=np.mean(y) - np.std(y), c="r", label="-1 desviacion estandar", linestyle="--")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.set_title("Variable Original")
ax.hist((y-np.mean(y))/np.std(y), bins=30)
ax.axvline(x = np.mean((np.mean(y-np.mean(y))/np.std(y))), c = "k", label="Media", linestyle="--")
ax.axvline(x = np.mean((np.mean(y-np.mean(y))/np.std(y))) + np.std((y-np.mean(y)))/np.std(y), c = "r", label="Media", linestyle="--")
```

```python
# Covarianza

fig, ax = plt.subplots()
ax.scatter(df["lar.petalo"], df["lar.sepalo"], alpha=0.7)
ax.set_xlabel("lar.petalo")
ax.set_ylabel("lar.sepalo")
ax.autoscale()
```

```python
np.cov(df["lar.petalo"], df["lar.sepalo"])
```

```python
corr = df.corr(method = "spearman")
```

```python
sns.heatmap(corr, xticklabels= corr.columns,yticklabels= corr.columns)
```

```python
mask = np.zeros_like(df.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr, 
            xticklabels= corr.columns,
            yticklabels= corr.columns, 
            cmap=sns.diverging_palette(20,220,n=200), 
            mask = mask, 
            annot = True, 
            center = 0)
```
